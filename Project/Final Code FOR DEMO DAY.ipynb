```python
import pandas as pd
import requests
import json
import webbrowser
from IPython.display import display,HTML


try:

    #Opens web page for trailer
    def trailer(title, year = ''):
        # run the removeSpaces function to make title

        # create url to trailer search results
        year = '(' + year + ')'
        url = 'https://www.youtube.com/results?search_query=' + title + year + '+trailer'

        # opens url in a new tab
        webbrowser.open_new_tab(url)


    #Build Parameters 
    def buildParameters(title, year = ''):
        t = "t=" + title
        if(len(year) > 3):
            y = "y=" + year
            parameters = t + "&" + y
            return parameters
        return t


    #Build Url 
    def buildURL(parameters):
        basic_url = 'http://www.omdbapi.com/?'
        key = 'apikey=ee06fce6'
        url = basic_url + parameters + '&' + key
        return url

    # Remove Spaces 
    def remove_spaces(string):
        string.strip().title()
        while (string.find(" ") > -1):
            string = string.replace(" ", "+") 

        return string


    # make DataFrame
    def makeDF(url):
        response = requests.get(url)
        movie = response.json()
        movie_df = pd.DataFrame(movie)
        return movie_df
    
    
    def displayMovie(movie_df):
        print("\n\n%s is a(n) %s film directed by %s." % (movie_df['Title'][0], movie_df['Genre'][0].lower(), movie_df['Director'][0]))
        print("\nHere's a plot summary. NO SPOILERS:\n%s" % (movie_df['Plot'][0]))
        print("\nSome key actors in this film are:\n%s" % (movie_df['Actors'][0]))
        display(movie_df[['Title','Runtime','Writer','imdbRating', 'Production', 'Language']].head(1))



    # MAIN CODE!!!
    display(HTML("<span style='color:blue;font-size: 30px;font-weight:bold;text-decoration: underline blue;font-family:Bookman Old Style'>Welcome to MovieGeek: Your one-stop-shop for film info!</span>"))

    #"Welcome to MovieGeek, your one-stop-shop for film info!" # we should look up how to style this text

    while True:
        answer = input('\nWould you like to search a movie? (Type "Yes" or "No") ').strip().title()
        if(answer == 'No'): # end program
            break

        elif(answer == 'Yes'): 
            message = "" # removes initial messge, makes it so user doesn't see unnecessary intro message
            title = input("What movie would you like to learn about? ").strip().title()
            # use build parameters, and build url, display DataFrame of info

            title_edit = remove_spaces(title) # makes title fit for url 

            # used in buildParameters() for more accurate search
            year = input("Do you know what year this movie was made? (yyyy or 'no'): ").strip()

            # make parameters for url
            parameters = buildParameters(title_edit, year = year)

            # build url
            url = buildURL(parameters)

            # make dataFrame

            movie_df = makeDF(url)


            print("\n\nHere is information on the most relevant title we found....")
            displayMovie(movie_df)

            # direct user to trailer
            text = input('Interested in %s? Want to watch the trailer? (Type "Yes" or "No") ' %(movie_df['Title'][0])).strip().title()


            if(text == "Yes"):
                print("Linking you to the %s movie trailer...." %(movie_df['Title'][0]))
                # run trailer function -- trailer(movie_df['Title'], date = movie_df['Year'][0])

                trailer(movie_df['Title'][0], year = movie_df['Year'][0])



        else:
            print("I'm sorry, we did not understand your question. Try again.")

except requests.exceptions.RequestException:
    print("ERROR: Cannot connect to ", url)
except json.decoder.JSONDecodeError:
    print("ERROR: Cannot decode the response into json")
except ValueError:
    print("\n\nWe had trouble finding your movie:\nCheck your spelling or add the correct date\n\nSorry for the inconvenience.")
```
